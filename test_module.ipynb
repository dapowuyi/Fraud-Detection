```python
from load_dataset_module import loadDataset
import user_statistics_module as u_stat

def main():
    
    # Ask user for a user ID
    while True:
        user = input ("Please enter a user id: ")
    
    #Check input that it is int
        try:
            user = int(user)
        except:
            print("Please input a number")
            continue
     #check that input is in the list of user IDs
        if user < 21 or user > 30:
            print("Please input a number between 21 and 30")
            continue
        break
    
    #load the dictionary into a varaiable
    myDict = loadDataset()
    #convert back to string
    userId = str(user)
    
    #call the minMax function
    (minAmt, maxAmt) = u_stat.minMax(myDict, userId)
    print("The minimum amount for user", userId, "is:", minAmt, "and the maximum amount is:", maxAmt, "\n")
    
    #call the location centroid function
    (x_coord, y_coord) = u_stat.locationCentroid(myDict, userId)
    print("The location centroid for user ", userId, "is:", x_coord, y_coord, "\n")
    
    #Call the distance function
    tranDist = u_stat.distance(myDict, userId)
    print("The distance for user", userId, "is:", tranDist, "\n")
    
    #Call the standard deviation function
    std = u_stat.stndDev(myDict, userId)    
    print("The standard deviation for the transactions performed by user ", userId, "is:", round(std, 2), "\n")
    
    #Call the variance function
    variance = u_stat.variance(myDict, userId)
    print("The variance for the transactions performed by user ", userId, "is ", round(variance, 2), "\n")
    
    #call the fradulent function
    fraud = u_stat.fraudulent(myDict, userId)

#Call the main function
main()
```
